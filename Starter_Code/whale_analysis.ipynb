```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
# YOUR CODE HERE
sp500_history_df = pd.read_csv(sp500_history_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
sp500_history_df.sort_index(inplace=True)
sp500_history_df.head()
```

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500?

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
# YOUR CODE HERE
joined_df.plot(kind="box", figsize=(20,10), title="Daily Returns", grid=True)
```

```python
largest_spread = ''
smallest_spread = ''
for key, value in joined_df.items():
    if largest_spread == '' and smallest_spread == '':
        largest_spread = key
        smallest_spread = key
    elif (joined_df[key].max() - joined_df[key].min()) > (joined_df[largest_spread].max() - joined_df[largest_spread].min()):
        largest_spread = key
    elif (joined_df[key].max() - joined_df[key].min()) < (joined_df[smallest_spread].max() - joined_df[smallest_spread].min()):
        smallest_spread = key
print(largest_spread)
print(smallest_spread)
```

***Which box has the largest spread? Which has the smallest spread?***<br>
*Tiger Global Management LLC has the largest spread and Paulson & Co. Inc has the smallest spread.  Spread in this case is calculated as max value - min value.*

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
# Which portfolios are riskier than the S&P 500?
# YOUR CODE HERE
standard_dev = joined_df.std()
standard_dev
```

```python
# Determine which portfolios are riskier than the S&P 500
# YOUR CODE HERE
for key, value in standard_dev.items():
    if value > standard_dev["S&P 500"]:
        print(key)
```

***Which portfolios are riskier than the S&P 500?***<br>
*Tiger Global Management LLC and Berkshire Hathaway Inc are riskier than the S&P.*

```python
# Calculate the annualized standard deviation (252 trading days)
# YOUR CODE HERE
volatility = standard_dev * np.sqrt(252)
volatility
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
# YOUR CODE HERE
joined_df.rolling(window=21).std().plot(figsize=(20,10), xlim=(joined_df.index.min(),joined_df.index.max()),title="Rolling 21-day Standard Deviation")
```

***Does the risk increase for each of the portfolios at the same time risk increases in the S&P?***<br>
*It does not appear to be that closely correlated for every portfolio but the standard deviation of the portfolios generally goes up when the standard deviation of the S&P goes up.*

```python
# Construct a correlation table
# YOUR CODE HERE
correlation = joined_df.corr()
correlation
```

***Which returns most closely mimic the S&P?***<br>
*The Algo 2 portfolio most closely mimics the S&P 500 returns.*

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)
# YOUR CODE HERE
covariance = joined_df["BERKSHIRE HATHAWAY INC"].cov(joined_df["S&P 500"])
variance = joined_df["S&P 500"].var()
BRK_beta = covariance / variance
BRK_beta
```

```python
# Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. 
# Does the portfolio seem sensitive to movements in the S&P 500?
rolling_covariance = joined_df["BERKSHIRE HATHAWAY INC"].rolling(window=60).cov(joined_df["S&P 500"])
rolling_variance = joined_df["S&P 500"].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20,10), title="Rolling 60-day Beta of BRK", xlim=(joined_df.index.min(),joined_df.index.max()))
```

***Does the portfolio seem sensitive to movements in the S&P 500?***<br>
*Based on the correlation calculation, Berkshire Hathaway is somewhat sensitive to the S&P.  About 75% of the movement in Berkshire Hathaway can be explained by the S&P.  Based on the beta calculations, Berkshire Hathaway is more volatile than the S&P, by about 13% (1.1347 simple beta).*

```python
# Calculate a rolling window using the exponentially weighted moving average. 
# YOUR CODE HERE
joined_df.ewm(halflife=21).std().plot(figsize=(20,10), xlim=(joined_df.index.min(),joined_df.index.max()), title="Exponentially Weighted Average")
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
# YOUR CODE HERE
sharpe_ratios = (joined_df.mean() * 252) / (joined_df.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
# YOUR CODE HERE
sharpe_ratios.plot(kind="bar", figsize=(20,10))
```

On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales? Type your answer here:<br>
No! Only Algo 1 outperforms the market and the whales.
